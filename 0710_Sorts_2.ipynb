# **선택정렬**

```python
import matplotlib.pyplot as plt

# 함수 정의
def VisualizeSelSort(numbers):
  i, n = 0, len(numbers)
  while i < n:                                           # 최솟값 찾기
    min = i                                              # 처음에 세팅되는 최소값
    print(f'현재 최솟값 : {min}')
    for j in range(min + 1, n):
      if numbers[j] < numbers[min]:
        min = j                                          # 최솟값 갱신
        print(f'최솟값 : {min}')
    plt.bar(range(n), numbers)
    plt.pause(0.5)
    numbers[i], numbers[min] = numbers[min], numbers[i]  # 찾은 최솟값과 인덱스 맨 앞 값과 교환
    i += 1
  print(f'선택정렬 후 결과: {numbers}')
  plt.bar(range(n), numbers)
  plt.pause(0.7)
  plt.show()

# 시작 및 함수 호출
numbers = input("숫자들을 입력하세요. (공백으로 구분해주세요): ").split()
print(numbers)
numbers = [int(num) for num in numbers]                  #숫자의 형태로 변환하여 리스트에 저장
print(numbers)
VisualizeSelSort(numbers)
```

# **버블정렬**

```python
def BBLSort(arr):
  n = len(arr)
  i = 0
  while i < n-1: # 첫 번째 반복문 (while)
    for j in range(n-1-i): # 두 번째 반복문 (for)
      if arr[j] > arr[j+1]: # 비교
        arr[j], arr[j+1] = arr[j+1], arr[j] # 교환
      print(f'정렬 결과:{arr}') # 주석 처리
    i += 1
# 사용자로부터 입력 받기
numbers = input("숫자들을 입력하세요 (공백으로 구분): ").split()
numbers = [int(num) for num in numbers]
# 버블정렬
BBLSort(numbers)
# 정렬 결과 출력
print("정렬된 결과:", numbers)
```

# **삽입정렬**

```python
a = [40, 60, 70, 50, 10, 20, 30]

i, n = 0, len(a)
while i < n - 1:
  #print(f'i = {i}')
  for j in range(i, 0, -1):
    #print(f'j = {j}', end = " ")
    if a[j - 1] > a[j]:
      #print(f'변경 전: a[j - 1] = {a[j -1]}, a[j] = {a[j]}')
      a[j], a[j - 1] = a[j - 1], a[j]
      #print(f'변경 후: a[j - 1] = {a[j -1]}, a[j] = {a[j]}')
    else:
      print(f'a[j - 1] = {a[j -1]}, a[j] = {a[j]} 변동없음')
    print(f'단계: {a}')
  print("------")
  i += 1
```

```python
# 오름차순 정렬
arr = ["u", "a", "o", "e", "i"]
print(f'정렬 전 : {arr}')
print("*"*60)
i = 1
while i < len(arr): # 정렬 안된 부분 인덱스 증가
  for j in range(i, 0, -1): # 정렬된 부분 인덱스 감소
    if arr[j - 1] > arr[j]: # 정렬된 부분과 비교
      arr[j], arr[j - 1] = arr[j - 1], arr[j]  # 교환
      print(f'[단계별 교환 후 정렬 과정] : {arr}') # 정렬 과정 출력
    else:
      print(f'[단계별 교환 후 정렬 과정] : [{arr[j]}]와 [{arr[j - 1]}] 값 크기 변동 없음')
    print()
  i += 1
print('+'*60)
print(f'정렬 후 : {arr}') #정렬 후 리스트 출력
```

```python
# 내림차순 정렬
arr = ["u", "a", "o", "e", "i"]
print(f'정렬 전 : {arr}')
print("*"*60)
i = 1
while i < len(arr): # 정렬 안된 부분 인덱스 증가
  for j in range(i, 0, -1): # 정렬된 부분 인덱스 감소
    if arr[j - 1] < arr[j]: # 정렬된 부분과 비교
      arr[j], arr[j - 1] = arr[j - 1], arr[j]  # 교환
      print(f'[단계별 교환 후 정렬 과정] : {arr}') # 정렬 과정 출력
    else:
      print(f'[단계별 교환 후 정렬 과정] : [{arr[j]}]와 [{arr[j - 1]}] 값 크기 변동 없음')
    print()
  i += 1
print('+'*60)
print(f'정렬 후 : {arr}') #정렬 후 리스트 출력
```
